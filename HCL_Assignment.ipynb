```python
import cv2
import string
import pytesseract
from google.colab.patches import cv2_imshow
from google.colab import drive
import string
```

**Define the image paths**

```python
img_paths = [
    "/content/drive/MyDrive/HCL_TECH/Screenshot 2023-10-20 at 2.49.53 PM.png",
    "/content/drive/MyDrive/HCL_TECH/Screenshot 2023-10-26 at 11.35.07 AM.png",
    "/content/drive/MyDrive/HCL_TECH/Screenshot 2023-10-26 at 11.35.18 AM.png",
    "/content/drive/MyDrive/HCL_TECH/Screenshot 2023-10-26 at 11.35.31 AM.png",
    "/content/drive/MyDrive/HCL_TECH/image.png"
]
```

```python
output_file = "/content/drive/MyDrive/HCL_TECH/ocr_output.txt"
```

**Define function to draw boxes around individual characters**

```python
def draw_boxes_on_character(img):
    img_height, img_width, _ = img.shape
    conf = r'-c tessedit_char_whitelist=' + string.digits
    boxes = pytesseract.image_to_boxes(img, config=conf)

    for box in boxes.splitlines():
        box = box.split(" ")
        character = box[0]
        x, y, x2, y2 = map(int, box[1:5])
        cv2.rectangle(img, (x, img_height - y), (x2, img_height - y2), (0, 255, 0), 1)
        cv2.putText(img, character, (x, img_height - y2), cv2.FONT_HERSHEY_COMPLEX, 0.75, (0, 0, 255), 1)

    return img
```

**Define function to draw boxes around detected text**

```python
def draw_boxes_on_text(img):
    raw_data = pytesseract.image_to_data(img)
    for count, data in enumerate(raw_data.splitlines()):
        if count > 0:
            data = data.split()
            if len(data) == 12:
                x, y, w, h, content = map(int, data[6:11])
                cv2.rectangle(img, (x, y), (w + x, h + y), (0, 255, 0), 1)
                content = str(content)
                cv2.putText(img, content, (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 255), 1)

    return img
```

```python
# Open output file in write mode
with open(output_file, 'w') as f:
    # Iterate through each image path
    for img_path in img_paths:
        # Read the image
        img = cv2.imread(img_path)

        # Convert BGR to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Perform OCR
        texts = pytesseract.image_to_string(img)
        print(texts)

        # Write text to the output file
        f.write(f"Text from {img_path}:\n{texts}\n\n")

# Notify user about successful saving
print(f"OCR output saved to: {output_file}")
```
